The task of dimensionality reduction through Singular Value Decomposition (SVD) is a very common approach when dealing with datasets with a large number of features, which could be more or less representative of the data distribution.
SVD algorithms aim to find an eigendecomposition of a matrix $A$ in the form:$$A=U \Sigma V^T$$
For vast datasets, the matrix representing the data assumes a very distinctive form, with a number of rows $r$ much larger than the number of columns $c$. Matrices with this distinctive shape ($r >> c$) are usually called "tall-and-skinny".<br>
In this particular case the dataset can even reach a size of some TB, and local computation may become arduous.

This work presentes an implementation of the *Direct TSQR* algorithm for matrix decomposition[1], which performs QR factorization of "tall-and-skinny" matrices in a ditributed environment with a MapReduce architecture.<br>
The distributed cluster is built and implemented with **Dask**.<br>
QR decomposition finds a representation for matrix A as shown below: $$A = QR$$
Adapting the algorithm to perform SVD is easy and only takes a couple of extra computation steps.

As shown in [1], among all the different algorithms performing QR decomposition, *Direct TSQR* remains stable even for "ill-conditioned" input matrices, *i.e.* matrices which contain many correlations in their rows and columns.

## Direct TSQR

### Useful functions
This is a list of functions that will be recalled and described later on.

```python
def extract_Q(M):
    Q = M[:][0]
    return Q

def extract_R(M):
    R = M[:][1]
    return R

def split_rows(M, n, rows):
    R = list()
    for i in range(n):
        R.append(M[i*rows:(i+1)*rows,:])
    return R

def cond_num(A):
    return np.linalg.cond(A.flatten().compute())

def est(Q):
    return np.linalg.norm(np.dot(Q.T,Q)-np.identity(len(Q.T)))
```

The **Direct TSQR** is a direct QR decoposition, which follows three principal steps.  
The goal is to perform a SVD of a large dataset collected in a "tall-and-skinny" matrix $A$. Therefore, a slight modification of the original TSQR algorithm is needed, *i.e.* a change in the second step and the addition of a fourth one.

Having the partitioned dataset, the **first step** is to compute a single QR decomposition over it.  
The result is a composition of $Q_{1,j}$ and $R_{1,j}$ matrices, where *j* goes from 1 to the number of partitions.

It is useful to *persist* the $Q_{1,j}$ matrices, so the implemented function `extract_Q()`, which selects *Q* from the output of the QR decomposition, is mapped across the partitions. 

```python
## Apply numpy QR decomposition to A using Dask map
S1 = A.map(np.linalg.qr) 

## Save the Qj matrices into the machine storage using the perisist() function
Q1 = client.persist(S1.map(extract_Q))
```

The **second step** is composed of two sub-steps:  
- reduce the $R_{1,j}$ matrices from the previous step (using an implemented function `extract_R()`) and perform a QR decomposition over the non-distributed result, which gives $R$ and $Q_2$;
- apply a local SVD on $R$, which gives three matrices $U_1$,$S$,$V^T$.

```python
#extract R from the QR decomposition and compute it
R1_col = S1.map(extract_R).flatten().compute()
#perform a numpy QR decomposition over the computed R matrix
Q2, R = np.linalg.qr(R1_col)

#compute a numpy SVD over the R matrix
U1, S, Vt = np.linalg.svd(R)
```

The **third step** consists on a matrix product over the persisted $Q_{1,j}$ of the first step and $Q_2$.  
This is done through a map task, but $Q_2$ is a numpy array so it must be converted to a dask array. For this reason $Q_2$ is split (calling the function `split_rows()`) and converted with `dask.array.stack()`.<br>
The same *stack* function is used on $Q_1$ to convert it from dask bag to dask array.<br>
The output of this step is called $Q$.

```python
#convert Q1 and Q2 into dask array
Q2_p = da.stack(split_rows(Q2, n_partitions, 8), axis=0)
Q1_p = da.stack(list(Q1), axis = 0)

#perform a matrix product
Q = da.map_blocks(lambda a,b : a @ b, Q1_p, Q2_p)

#delete Q1 from the storage
del Q1
```

The **fourth step** consists in a matrix product between $Q$ and $U_1$. It returns a matrix which is the actual $U$ of the SVD over $A$.

```python
#matrix prodoct
U = da.map_blocks(lambda a,b : a @ b, Q, U1)
#computation of U
Compute=True
if Compute == True:
    #concatenate of the U partitioned matricies
    U = np.concatenate(U.compute(), axis=0)
```

All the steps are collected inside a single python function.<br> 

```python
def parallel_SVD(A, n_partitions, Compute=True, Accuracy=False):
    
    #First step
    S1 = A.map(np.linalg.qr)
    Q1 = client.persist(S1.map(extract_Q)) # save Q1 in memory becouse we'll need it later
    
    #Second step
    R1_col = S1.map(extract_R).flatten().compute()
    Q2, R = np.linalg.qr(R1_col)
    
    #Second step for for SVD
    U1, S, Vt = np.linalg.svd(R)
    
    #Third step
    Q2_p = da.stack(split_rows(Q2, n_partitions, len(R)), axis=0)
    Q1_p = da.stack(list(Q1), axis = 0)
    Q = da.map_blocks(lambda a,b : a @ b, Q1_p, Q2_p)
     
    del Q1
    
    #Fourth step (SVD only)
    U = da.map_blocks(lambda a,b : a @ b, Q, U1)
    
    if Compute == True:
        U = np.concatenate(U.compute(), axis=0)
    if Accuracy == True:
        print("Accuracy Q: ",est(np.concatenate(Q.compute(), axis=0)))
        print("Accuracy U: ",est(U))
        
    return(U,S,Vt)
```

The Accuracy parameter enables the accuracy evaluation. This is done by checking the orthogonality of $Q$ through the following *error function*:

$$ ||Q^{T}Q - \mathbb{1}||_2$$

# Parallel HAVOK Analysis of a Lorenz System
New libraries imported:

```python
import pysindy as ps
import scipy as sp
import control
import control.matlab
from matplotlib import cm
from matplotlib import colors
```

## HAVOK - Hankel Alternative View Of Koopman

HAVOK consists of a universal, data-driven decomposition of chaos as an intermittently forced linear system, performed with a combination of delay embedding and Koopman theory.

The algorithm is able to build a linear representation of a multi-dimensional chaotic system, starting from a unidimesional time series of the most meaningful coordinate.<br>
This is possible by building a *Hankel matrix* of the time series and performing SVD to select the first $r$ eigenvectors as the training set for a sparse regression. The first $r-1$ vectors are taken as the linear coordinates of the model, while the $r^{th}$ represents the forcing term that influences the dynamics.<br>
A detailed explanation of the algorithm and the theory behind it is described in [2].

In this work, HAVOK analysis is performed on the Lorenz attractor, a well known example of chaotic dynamics.

## Code implemented

### Lorenz data creation

```python
def lorenz(t,y,sig,rho,beta):
    dx = sig*(y[1] - y[0])
    dy = y[0]*(rho - y[2]) - y[1]
    dz = y[0]*y[1] - beta*y[2]
    return [dx,dy,dz]

solver = sp.integrate.solve_ivp(fun=lambda t, y: lorenz(t,y,10,28,8/3),
                                 t_span=[0,1500.1], t_eval=np.arange(0,1500.1,0.001), y0=[-8,8,27], method='LSODA', 
                                 dense_output=False,rtol=1e-18,atol=1e-22)

data_ts = solver.y.T
data_ts = data_ts[:,0]
```

```python
n_partitions = 6
```

```python
# Saving into common folder

l = 100
len_data = len(data_ts) - l
for i in range(n_partitions):
    data_p = data_ts[i*int(len_data/n_partitions):(i+1)*int(len_data/n_partitions)+100]
    data_p.tofile('DATASET/Lorenz/data_ts_{}.csv'.format(i+1), sep = ' ')
```

### Train import

```python
n_train = n_partitions

x_b = db.from_sequence(['DATASET/Lorenz/data_ts_{}.csv'.format(i+1) for i in range(n_train)]).map(np.loadtxt)
x_b
```

### Hankel matrix
The first step of the HAVOK analysis consists of the division of the time series into windows of length `l` and the creation of an Hankel matrix `H`, stacking the windows by rows. Since data is partitioned, it is convenient to create an Hankel matrix for every chunk with the condition that the components of each $H_i$ are smoothly connected, *i.e.* the last row of $H_i$ must repeat most of the values contained in the first row of $H_{i+1}$. In this way the concatenation of all of the $H_i$ results into a "tall-and-skinny" Hankel matrix.<br>
In order to have a simpler algorithm, data has been saved in advance in a way that this condition is verified (the last `l` data contained in each partition are the first data of the subsequent one).

```python
def Hankel(x_ts,l):
    l=100
    H = np.zeros((len(x_ts),l))
    for i in range(len(x_ts)-l): H[i,:] = x_ts[i:l+i]
    return H

l = 100
H = x_b.map(lambda x: Hankel(x,l))
```

### Parallel SVD
`H` is a "tall-and-skinny" matrix so we can perform the parallel SVD through *Direct TSQR*. In this case `U` is a matrix with the same dimensions of `H` but in order to perform the HAVOK analysis only the first `r` columns are needed.

```python
U,S,Vt = parallel_SVD(H, n_partitions=n_train, Compute=False)

r = 15
u = np.concatenate(np.array(U[:,:,:r]),axis=0)
```

### Sparse Regression
The sparse regression is performed locally using a method from `pySindy` to build the model:

$$dU = AU_{[1,r-1]} + BU_r$$

```python
dt = 0.001
time = np.arange(0,len(u)*dt,dt)

differentiation_method = ps.FiniteDifference(order=2)
feature_library = ps.PolynomialLibrary(degree=1)
optimizer = ps.STLSQ(threshold=0.02)

model = ps.SINDy(
    differentiation_method=differentiation_method,
    feature_library=feature_library,
    optimizer=optimizer
)

model.fit(u,t=time)

M = model.coefficients()
M = M[:,1:r+1]
A = M[:r-1,:r-1]
B = M[:r-1,r-1]

df = pd.DataFrame(np.round(M))
df
```

```python
fig, ax = plt.subplots(figsize = (6,6))

ax.set_title("A matrix")
divnorm=colors.TwoSlopeNorm(vmin=np.min(A), vcenter=0., vmax=np.max(A))
cax = ax.matshow(A, cmap= plt.get_cmap('PiYG'), norm=divnorm)
fig.colorbar(cax)
```

### Integration of the linear model
Using `A` and `B` and knowing the forcing term, the linear model can be integrated with the `control.StateSpace` method and compared with the embedded time series.<br>
This is basically a numerical integration of a time series, so the computation can be done separately for each partition using the first row of each $U_i$ as initial conditions.

```python
B = np.reshape(B,(r-1,1)) # control.StateSpace requires B with shape (r-1,1)
sys = control.StateSpace(A,B,np.eye(r-1),0*B)
Ur = db.from_sequence(np.array(U[:,:,:r]))

# the function lsim return something of the kind (y_out,time,...)
def extract_Y(M):
    Y = M[:][0]
    return Y

result = Ur.map(lambda x: control.matlab.lsim(sys,x[:,r-1],np.arange(0,len(x)*dt,dt),x[0,:r-1]))\
            .map(extract_Y).compute()

y = np.concatenate(result,axis=0)
```

As one can see, the linear model recreates the behaviour of the embedded time series. It is also evident that there is a correlation between the peaks of the forcing term and the switching of the lobe.

```python
fig, ax = plt.subplots(2,1, figsize = (16,8))
ax[0].plot(time[:50000],u[:50000,0],c='black',label='embedded time series')
ax[0].plot(time[:50000],y[:50000,0],c='red',label='linear model',lw=1)
ax[0].set_ylabel('u1',rotation=0)
ax[0].legend(loc='upper right')
ax[1].plot(time[:50000],u[:50000,-1],c='black',lw=1)
ax[1].set_ylabel('f',rotation=0)
ax[1].set_xlabel('time [s]')
```

Though the same analysis performed locally wields much better results. As shown in the graph below, the integrated time series follows the original time series more truthfully.<br>

## Possible developments
Since `A` and `B` are constant, once they have been computed using a training set, they can be used to integrate the rest of the dataset to make predictions about the lobe switching. Still, the integration requires the advance knowledge of the forcing term but this can be calculated by convoluting `V` with a slinding window of length `l` of the incoming data.
By setting a threshold in the forcing term it is possible to predict shortly in advance the chaotic behaviour of the Lorenz system *i.e.* the lobe switching.



## Conclusions
The analysis on the California Housing Dataset shows that the *Direct TSQR* algorithm indeed has a good accuracy in the decomposition of "tall-and-skinny" matrices.<br>
By running it on a larger dataset and in a more complex context it was possible to see how the time performances get better with a higher parallelization.<br>
On the other hand, when comparing the reconstructed time series, the results from the parallel computation are less optimal than the "direct" ones. This may be due to the fact that the whole computation is more complicated, and the errors propagation has a more dominant impact, expecially in this particular case, since the system dynamic is chaotic, *i.e.* highly influenced by small perturbations.
